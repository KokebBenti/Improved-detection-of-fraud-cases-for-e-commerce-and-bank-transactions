**Merge Datasets for Geolocation Analysis**

```python
df2.info()
```

```python
#Convert IP addresses to integer format
df2=EDA.to_int(df2,"lower_bound_ip_address")
df2=EDA.to_int(df2,"upper_bound_ip_address")
```

```python
#merge The two dataframes
interval_index = pd.IntervalIndex.from_arrays(df2['lower_bound_ip_address'], df2['upper_bound_ip_address'], closed='both')
matched_idx = interval_index.get_indexer(df1['ip_address'])
df2_matched = pd.DataFrame(index=df1.index, columns=df2.columns)
df2_matched.loc[matched_idx != -1] = df2.iloc[matched_idx[matched_idx != -1]].values

merged = pd.concat([df1, df2_matched], axis=1)
print(merged)
```

**Feature Engineering**

```python
merged["time_since_signup"]=(merged['purchase_time']-merged['signup_time']).dt.total_seconds()
```

```python
merged['hour_of_day']=merged['purchase_time'].dt.hour
merged['Day_of_week']=merged['purchase_time'].dt.day_of_week
```

```python
merged=merged.sort_values(['user_id', 'purchase_time'])
merged=merged.groupby('user_id', group_keys=False).apply(EDA.create_features)
```
